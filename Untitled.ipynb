```python
# Butterworth bandpass filter function
def butter_bandpass_filter(data, lowcut, highcut, fs, order=2):
    from scipy.signal import butter, lfilter
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    y = lfilter(b, a, data)
    return y

# Function to preprocess the audio files
def preprocess_audio(file_path, sr=1000, duration=2.5):
    data, sr = librosa.load(file_path, sr=sr)
    data = butter_bandpass_filter(data, 25, 400, sr)
    data = librosa.util.fix_length(data, int(sr * duration))
    return data

# Load and preprocess dataset
def load_and_preprocess_dataset(file_paths):
    dataset = []
    labels = []
    for file_path, label in file_paths:
        data = preprocess_audio(file_path)
        dataset.append(data)
        labels.append(label)
    return np.array(dataset), np.array(labels)
```

```python
import glob
import numpy as np
import scipy.signal as signal
import librosa
import soundfile as sf
import os
```

```python
def preprocess_audio(file_path, lowcut=25, highcut=400, fs=1000, duration=2.5):
    # Load audio file
    y, sr = librosa.load(file_path, sr=fs)
    
    # Bandpass filter
    nyquist = 0.5 * sr
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = signal.butter(2, [low, high], btype='band')
    y_filtered = signal.filtfilt(b, a, y)
    
    # Truncate or pad the signal to the specified duration
    max_len = int(duration * sr)
    if len(y_filtered) > max_len:
        y_truncated = y_filtered[:max_len]
    else:
        y_truncated = np.pad(y_filtered, (0, max_len - len(y_filtered)), 'constant')
    
    # Normalize the signal to the range (-1, 1)
    y_normalized = y_truncated / np.max(np.abs(y_truncated))
    
    return y_normalized
```

```python
import librosa
import numpy as np

def time_stretch(y, rate=1.2):
    return librosa.effects.time_stretch(y, rate)

def pitch_shift(y, sr, n_steps=4):
    return librosa.effects.pitch_shift(y, sr, n_steps)

def add_noise(y, noise_factor=0.005):
    noise = np.random.randn(len(y))
    augmented_data = y + noise_factor * noise
    return augmented_data

def augment_data(y, sr):
    augmented_data = []
    augmented_data.append(time_stretch(y, rate=0.8))
    augmented_data.append(time_stretch(y, rate=1.2))
    augmented_data.append(pitch_shift(y, sr, n_steps=-2))
    augmented_data.append(pitch_shift(y, sr, n_steps=2))
    augmented_data.append(add_noise(y, noise_factor=0.005))
    augmented_data.append(add_noise(y, noise_factor=0.01))
    return augmented_data
```

```python
D = librosa.stft(y)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
S_db.shape
```

```python
# Plot the transformed audio data
fig, ax = plt.subplots(figsize=(10, 5))
img = librosa.display.specshow(S_db,
                              x_axis='time',
                              y_axis='log',
                              ax=ax)
ax.set_title('Spectogram Example', fontsize=20)
fig.colorbar(img, ax=ax, format=f'%0.2f')
plt.show()
```

```python
S = librosa.feature.melspectrogram(y=y,
                                   sr=sr,
                                   n_mels=128 * 2,)
S_db_mel = librosa.amplitude_to_db(S, ref=np.max)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
# Plot the mel spectogram
img = librosa.display.specshow(S_db_mel,
                              x_axis='time',
                              y_axis='log',
                              ax=ax)
ax.set_title('Mel Spectogram Example', fontsize=20)
fig.colorbar(img, ax=ax, format=f'%0.2f')
plt.show()
```

# Model

```python
import os
import librosa
import numpy as np
import glob

def extract_features(file_path, sr=22050, n_mfcc=13):
    y, sr = librosa.load(file_path, sr=sr)
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    return np.mean(mfcc.T, axis=0)

data_dir = '/Users/koyalchandresh/Desktop/Ausa/Steth-main/database/Audio Files'  # Update this path

# Using glob to match files
file_pattern = os.path.join(data_dir, '*.wav')

X = []
file_paths = []

for file_path in glob.glob(file_pattern):
    try:
        features = extract_features(file_path)
        X.append(features)
        file_paths.append(file_path)
    except Exception as e:
        print(f"Error processing {file_path}: {e}")

X = np.array(X)

print(f"Extracted features from {len(X)} audio files.")
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

n_clusters = 6  # Adjust based on the number of classes

kmeans = KMeans(n_clusters=n_clusters, random_state=42)
clusters = kmeans.fit_predict(X)

# Visualize clusters
plt.scatter(X[:, 0], X[:, 1], c=clusters)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-means Clustering of Audio Features')
plt.show()
```
